```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

Primera se√±al para el filtro, CH1 entrada, CH2 salida.

```python
data7 = pd.read_csv("NewFile7.csv", skiprows=1)
data7.drop(data7.columns[3], axis=1, inplace=True)
t = data7.iloc[:,0]*1000
v1 = data7.iloc[:,1]
v2 = data7.iloc[:,2]
# rojo es CH1, azul es CH2
# en las fotos CH1 es amarillo pero no se ve bien con fondo blanco
plt.plot(t,v1,color="red",label="CH1")
plt.plot(t,v2,label="CH2")
plt.gca().xaxis.set_major_locator(MultipleLocator(5))
plt.xlabel("Tiempo (ms)")
plt.ylabel("Voltaje (V)")
plt.title("Senoidal 90Hz, 5Vpp, Offset=0, DC")
plt.legend()
plt.grid(True)
plt.show()
```

Ahora se aument√≥ la frecuencia de la se√±al de entrada a 90 KHz

```python
data8 = pd.read_csv("NewFile8.csv", skiprows=1)
data8.drop(data8.columns[3], axis=1, inplace=True)
t = data8.iloc[:,0]*1000000
v1 = data8.iloc[:,1]
v2 = data8.iloc[:,2]
# rojo es CH1, azul es CH2
plt.plot(t,v1,color="red",label="CH1")
plt.plot(t,v2,label="CH2")
plt.gca().xaxis.set_major_locator(MultipleLocator(5))
plt.xlabel("Tiempo (us)")
plt.ylabel("Voltaje (V)")
plt.title("Senoidal 90KHz, 5Vpp, Offset=0, DC")
plt.legend()
plt.grid(True)
plt.show()
```
