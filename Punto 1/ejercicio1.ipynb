```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import numpy as np
```

```python
data0 = pd.read_csv("NewFile0.csv", skiprows=1)
data0.drop(data0.columns[2], axis=1, inplace=True)
```

```python
# rojo es CH1, azul es CH2
# en las fotos CH1 es amarillo pero no se ve bien con fondo blanco
plt.plot(data0.iloc[:,0]*1000, data0.iloc[:,1]/10,color="red")
plt.gca().xaxis.set_major_locator(MultipleLocator(0.25))
plt.xlabel("Tiempo (ms)")
plt.ylabel("Voltaje (V)")
plt.title("Senoidal 2KHz, 5Vpp, Offset=0, DC")
plt.grid(True)
plt.show()
```

```python
data1 = pd.read_csv("NewFile1.csv", skiprows=1)
data1.drop(data1.columns[2], axis=1, inplace=True)
plt.plot(data1.iloc[:,0]*1000000, (data1.iloc[:,1]/10)*1000,color="red")
plt.gca().xaxis.set_major_locator(MultipleLocator(25))
plt.xlabel("Tiempo (us)");
plt.ylabel("Voltaje (mV)");
plt.title("Senoidal 25KHz, 0.4Vpp, Offset=0.2V, DC")
plt.grid(True)
plt.show()
```
